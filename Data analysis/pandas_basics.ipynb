# Dropping

- to drop columns
     - df.drop() -> removes columns
- to drop rows
     - df.drop() -> removes rows
     - df.dropna() -> removes rows with NaN
     - df.drop_duplicates() -> removes row with duplicate values

```python
cols_to_drop = ['AREA', 'REG', 'DEV', 'Type', 'Coverage', 'Faltu', 'another column']
df.drop(columns=cols_to_drop, inplace=True) # a inplace true leads to dropppng in original data frame rather than a copy    
```

```python
df.drop_duplicates(subset=['AreaName'])
```
